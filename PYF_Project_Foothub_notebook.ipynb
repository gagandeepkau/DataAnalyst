### **Question 13:** The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer. [3 marks]

```python
df_rated = df[df['rating'] != 'Not given'].copy()
# Convert rating column from object to integer
df_rated['rating'] = df_rated['rating'].astype('int')
df_rat_cnt = df_rated.groupby(['restaurant_name'])['rating'].count().sort_values(ascending = False).reset_index()
df_rat_cnt_50 = df_rat_cnt[df_rat_cnt['rating']>50]
rest_names = df_rat_cnt_50.groupby(['restaurant_name'])['rating'].mean()>4
rest_names.reset_index()

```

#### Observations: Only 4 restaurants 'Blue Ribbon Fried Chicken', 'Blue Ribbbon Sushi', 'Shake Shack' and 'The Meatball Shop' meet the above criteria of have a rating count of more than 50 and the average rating greater than 4.


### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python


df['revenue'] = [i*0.25 if i > 20 else i*0.15 if i > 5 else i *0 for i in df.cost_of_the_order]
total_rev = df['revenue'].sum()
print('The net revenue is around',round(total_rev,2), 'dollars')
```

#### Observations: The reveue is generated more when order costs more than 20$.


### **Question 15:** The company wants to analyze the total time required to deliver the food. What percentage of orders take more than 60 minutes to get delivered from the time the order is placed? (The food has to be prepared and then delivered.) [2 marks]

```python
# Write the code here
```

```python

df['total_time'] = df['food_preparation_time'] + df['delivery_time']
order_above_60 = df[df['total_time'] > 60]
order_above_60.shape[0]
percent_above_60 = round(order_above_60.shape[0]/df.shape[0]* 100,2)
print('The total number of orders that took more than 60 minutes to get delivered is',percent_above_60,'%')
```

#### Observations:About 10.54% of customers wait for their order for more than 1 hour. This can be area of improvement and need to be montiored on continous basis.


### **Question 16:** The company wants to analyze the delivery time of the orders on weekdays and weekends. How does the mean delivery time vary during weekdays and weekends? [2 marks]

```python

df_day_weekend = df[df['day_of_the_week'] =='Weekend'].copy()
print('The mean delivery time during weekend',round(df_day_weekend['delivery_time'].mean(),2),'mins')
df_day_weekday = df[df['day_of_the_week'] =='Weekday'].copy()
print('The mean delivery time during weekday',round(df_day_weekday['delivery_time'].mean(),2),'mins')
```

#### Observations:
The mean delivery time during weekday is more as compared to mean delivery time on weekend. This could be due to various factors like traffic during weekdays, availability of more staff on weekends.
